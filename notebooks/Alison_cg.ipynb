**Data Exploration**

```python
#merging tables to be able to run linear models in the data table. 
X_train_mastery = pd.merge(X_train, champion_mastery, how='left', on=['summonerId', 'championId'], indicator=True)
X_train_mastery
```

```python
mastery_predictions = ( 
    X_train_mastery
    .sort_values(['matchId', 'championPoints'], ascending = [True, True])
    .drop_duplicates('matchId')
    [['matchId', 'championPoints']]
    .reset_index(drop = True)
)

mastery_predictions
```

```python
from sklearn.metrics import accuracy_score

mastery = accuracy_score(
    y_true = y_train['matchId'],
    y_pred = mastery_predictions['championPoints']
)
```

```python
X_test_mastery = pd.merge(X_test, champion_mastery, how='left', on=['summonerId', 'championId'], indicator=True)
X_test_mastery
```

```python
mastery_test = ( 
    X_test_mastery
    .sort_values(['matchId', 'championPoints'], ascending = [True, True])
    .drop_duplicates('matchId')
    [['matchId', 'championPoints']]
    .reset_index(drop = True)
)

mastery_test
```

```python
y_pred = ( 
     X_test_mastery
    .sort_values(['matchId', 'championPoints'], ascending = [True, False])
    .drop_duplicates('matchId')
    ['championPoints']
    .reset_index(drop = True)
)

y_pred
```

```python
submission['winner'] = y_pred
submission.head()
```
